# Salary Prediction using Linear Regression

```python
import pandas as pd 
from matplotlib import pyplot as plt 
```

```python
df = pd.read_csv("Salary Data.csv", sep=',')
df.head()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Experience Years'],df.Salary)
plt.xlabel("Experience Years",fontsize=20)
plt.ylabel("Salary",fontsize=20)
plt.show()
```

```python
xmean=df['Experience Years'].mean()
ymean=df['Salary'].mean()
```

```python
data=df[['Experience Years','Salary']]
data
```

```python
plt.figure(figsize=(10,6))
plt.scatter(xmean,ymean,color='r')
plt.scatter(df['Experience Years'],df.Salary)
plt.xlabel("Experience Years",fontsize=20)
plt.ylabel("Salary",fontsize=20)
plt.show()
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
reg
```

```python
xfit=df[['Experience Years']]
yfit=df[['Salary']]
```

```python
reg.fit(df[['Experience Years']],df[['Salary']]) #use DataFrame
```

```python
m=reg.coef_
c=reg.intercept_
```

```python
m*2.1+c
```

```python
df['predict_y']=reg.predict(df[['Experience Years']]) #alwas using DataFrame
```

```python
df
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['Experience Years'],df['predict_y'])
plt.scatter(xmean,ymean,color='r')
plt.scatter(df['Experience Years'],df.Salary)
plt.xlabel("Experience Years",fontsize=20)
plt.ylabel("Salary",fontsize=20)
plt.show()
```

# Cost & Lose

```python
df['loss']=df['Salary']-df['predict_y']
```
